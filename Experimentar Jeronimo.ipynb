### Efecto del estrato en las áreas distintas

```python
columna_estrato = 'FAMI_ESTRATOVIVIENDA'
estratos = ['Sin Estrato', 'Estrato 1', 'Estrato 2', 'Estrato 3', 'Estrato 4', 'Estrato 5', 'Estrato 6']
areas = ['PUNT_MATEMATICAS', 'PUNT_INGLES', 'PUNT_SOCIALES_CIUDADANAS', 'PUNT_C_NATURALES', 'PUNT_LECTURA_CRITICA']
labels_areas =["Puntaje en matemáticas", "Puntaje en Inglés", "Puntaje en ciencias sociales", "Puntaje en ciencias naturales", "Puntaje en lectura crítica"]

colores = {
    "PUNT_MATEMATICAS": "#42fff9",
    "PUNT_INGLES": "#fc9840",
    "PUNT_SOCIALES_CIUDADANAS": "#4eff4e",
    "PUNT_C_NATURALES": "#ff714a",
    "PUNT_LECTURA_CRITICA": "#b15fff"
}

means = df.groupby(columna_estrato)[areas].mean()
means = means.reindex(estratos)

x = np.arange(len(means.index))
n = len(areas)
group_w = 0.65
w = group_w / n


fig, ax = plt.subplots(figsize=(14, 5))

for i, area in enumerate(areas):
    ax.bar(x+(i-(len(areas)-1)/2)*w, means[area].values, width=w*0.85, label = labels_areas[i], color= colores[area])

ax.set_xticks(x)
ax.set_xticklabels(means.index.astype(str))


ax.set_title('Áreas más afectadas por condición económica')
ax.set_xlabel('Estrato')
ax.set_ylabel('Puntaje promedio')
ax.legend(ncol=len(areas), bbox_to_anchor=(0.5, -0.15), loc='upper center', frameon=False)

plt.tight_layout()
plt.show()


```

### Acceso a internet y desempeño en idiomas

```python
df["DESEMP_INGLES"].value_counts()
```

```python
internet_columna = 'FAMI_TIENEINTERNET'
columna_nivel_ing = 'DESEMP_INGLES'
columna_ingles = "PUNT_INGLES"
columna_global = "PUNT_GLOBAL"

niveles = ["A-", "A1", "A2", "B1", "B+"]

internet = df[internet_columna].astype(str).str.lower().str.strip()
con_internet = internet.isin(["si"])

#Grafico 1

df2=df.copy()
df2["Internet"] = np.where(con_internet, "Con internet", "Sin internet")

print(df2["Internet"].value_counts())

fig, axes = plt.subplots(1, 2, figsize=(16, 6))
fig.suptitle('Acceso a internet y desempeño en idiomas', fontsize=16, fontweight='bold')

ax1 = axes[0]

conteos = (df2.groupby([columna_nivel_ing, "Internet"]).size().unstack(fill_value=0))

conteos = conteos.reindex(niveles)

x = np.arange(len(conteos.index))
w = 0.5

ax1.bar(x - w/2, conteos.get("Con internet", 0), width = w*0.7, label = "Con internet", color = "#03eaf2")
ax1.bar(x + w/2, conteos.get("Sin internet", 0), width = w*0.7, label = "Sin internet", color = '#ff5a3d')
        
ax1.set_title('Distribución de niveles de ingles por acceso a internet')
ax1.set_xticks(x)
ax1.set_xticklabels(conteos.index)
ax1.set_ylabel('Cantidad de estudiantes')
ax1.legend(frameon=False)

#Grafico 2
ax2 = axes[1]


m = df2[[columna_ingles, columna_global, 'Internet']].dropna()
for grp, col in [('Con internet', "#03eaf2"), ('Sin internet', '#ff5a3d')]:
    sub = m[m['Internet'] == grp]
    ax2.scatter(sub[columna_ingles], sub[columna_global], s=30, alpha=0.45, label=grp, c=col)

ax2.set_title('Puntaje inglés vs puntaje global')
ax2.set_xlabel('Punt. Inglés')
ax2.set_ylabel('Punt. Global')
ax2.legend(frameon=False)

plt.tight_layout()
plt.show()



```

### Dispersión de puntajes por tipo de colegio y por zona

```python
df["COLE_NATURALEZA"]
```

```python
tipo_colegio = "COLE_NATURALEZA"
zona_colegio = "COLE_AREA_UBICACION"

puntaje = "PUNT_GLOBAL" # PUNT_MATEMATICAS - PUNT_INGLES - PUNT_SOCIALES_CIUDADANAS PUNT_C_NATURALES - PUNT_LECTURA_CRITICA - PUNT_GLOBAL

areas =["PUNT_MATEMATICAS", "PUNT_INGLES", "PUNT_SOCIALES_CIUDADANAS", "PUNT_C_NATURALES", "PUNT_LECTURA_CRITICA"]
temporal = df[[tipo_colegio, zona_colegio]+ areas].dropna().copy()
temporal[tipo_colegio] = temporal[tipo_colegio].astype(str).str.lower().str.strip()
temporal[zona_colegio] = temporal[zona_colegio].astype(str).str.lower().str.strip()

def mapa_tipo(colegio):
    if "no oficial" in colegio:
        return "Privado"
    return "Oficial"

def mapa_zona(colegio):
    if "rural" in colegio:
        return "Rural"
    return "Urbano"

temporal["Tipo"] = temporal[tipo_colegio].map(mapa_tipo)
temporal["Zona"] = temporal[zona_colegio].map(mapa_zona)
temporal["Grupo"] = temporal["Tipo"] + " - " + temporal["Zona"]

temporal.drop(tipo_colegio, axis=1,inplace=True)
temporal.drop(zona_colegio, axis=1, inplace=True)

orden = ["Oficial - Urbano", "Oficial - Rural", "Privado - Urbano", "Privado - Rural"]

colores = colores = {
    "Oficial - Urbano": "#42fff9",
    "Oficial - Rural": "#fc9840",
    "Privado - Urbano": "#4eff4e",
    "Privado - Rural": "#b15fff",
}


tabla = (temporal.groupby("Grupo")[areas].mean().reindex(orden))
tabla = tabla.T
 

import plotly.express as px

fig = px.imshow(tabla,
                text_auto = ".1f",
                aspect = "equal",
                labels=dict(x="Grupo", y="Área", color= '#RdYlGn'),
                x=["Oficial - Urbano", "Oficial - Rural", "Privado - Urbano", "Privado - Rural"],
                y=["Matemáticas", "Inglés", "Ciencias Sociales", "Ciencias Naturales", "Lectura Crítica"]
               )
fig.update_xaxes(side="top")
fig.update_layout(
    title="Puntaje promedio por grupo y área",
    width=700,                          # ✅ ajusta tamaño (prueba 650–850)
    height=800,                         # ✅ casi cuadrado
    margin=dict(l=80, r=40, t=80, b=60),# ✅ menos blanco
    coloraxis_colorbar=dict(
        title="Puntaje Promedio",
        len=0.65,                       # altura barra color
        thickness=30,                   # grosor
        x=1.3                       # pegada al heatmap
    )
)



fig.show()



```

### Tipo de documento de los estudiantes

```python

conteo = df["ESTU_TIPODOCUMENTO"].value_counts()

labels = conteo.index.tolist()
sizes  = conteo.values.tolist()

top_labels = set(labels[:3])


def top3(pct):
    top3.i += 1
    label = labels[top3.i - 1]
    return f"{pct:.1f}%" if label in top_labels else ""
top3.i = 0

fig, ax = plt.subplots(figsize=(7, 5))

wedges, texts, autotexts = ax.pie(
    sizes,
    labels=None,
    autopct=top3,
    startangle=90
)


ax.set_title("Distribución por tipo de documento")
ax.axis("equal")

legend_labels = [f"{lab} ({val})" for lab, val in zip(labels, sizes)]
ax.legend(
    wedges,
    legend_labels,
    title="Tipos",
    loc="center left",
    bbox_to_anchor=(1.02, 0.5)
)

plt.tight_layout()
plt.show()
```

### Ubicación del colegio (rural y urbano)

```python
columna = "COLE_AREA_UBICACION"
col = df[columna].astype(str).str.strip().str.lower()

counts = col.value_counts(dropna=False)
prop = col.value_counts(normalize=True, dropna=False) * 100

plt.figure(figsize=(6,4))
ax = prop.plot(kind="bar")

plt.title("Ubicación del colegio (porcentaje)")
plt.xlabel("Categoría")
plt.ylabel("Porcentaje (%)")
plt.xticks(rotation=0)

for i, categoria in enumerate(prop.index):
    porcentaje = prop.loc[categoria]
    n = counts.loc[categoria]
    ax.text(
        i, porcentaje,
        f"{n} ({porcentaje:.1f}%)",
        ha="center", va="bottom",
        fontsize=9
    )

ax.set_ylim(0, prop.max() * 1.15)

plt.tight_layout()
plt.show()

```

### Municipio de residencia del estudiante

```python
df["ESTU_MCPIO_RESIDE"].value_counts()
```

```python
columna = "ESTU_MCPIO_RESIDE"

col = (df[columna]
       .astype("string")
       .str.strip()
       .str.upper()
       .fillna("SIN INFORMACIÓN")
)

counts = col.value_counts()
total = counts.sum()

top_n = 5
top = counts.head(top_n)
otros = total - top.sum()

counts_plot = top.copy()
if otros > 0:
    counts_plot["OTROS"] = otros

prop_plot = counts_plot / total * 100

plt.figure(figsize=(10,5))
ax = prop_plot.plot(kind="bar")

plt.title(f"Municipio de residencia (Top {top_n} + Otros) - Porcentaje")
plt.xlabel("Municipio")
plt.ylabel("Porcentaje (%)")
plt.xticks(rotation=45, ha="right")

for i, cat in enumerate(prop_plot.index):
    pct = prop_plot.loc[cat]
    n = counts_plot.loc[cat]
    ax.text(i, pct, f"{n} ({pct:.1f}%)", ha="center", va="bottom", fontsize=8)

ax.set_ylim(0, prop_plot.max() * 1.2 if prop_plot.max() > 0 else 1)
plt.tight_layout()
plt.show()
```

### Personas en el hogar

```python
df["FAMI_PERSONASHOGAR"].value_counts().sort_index()
```

```python
columna = "FAMI_PERSONASHOGAR"

col = (df[columna]
       .astype("string")
       .str.strip()
       .str.upper()
       .fillna("SIN INFORMACIÓN")
)

mapa_numeros = {
    "UNA": 1, "UNO": 1,
    "DOS": 2,
    "TRES": 3,
    "CUATRO": 4,
    "CINCO": 5,
    "SEIS": 6,
    "SIETE": 7,
    "OCHO": 8,
    "NUEVE": 9,
    "DIEZ": 10,
    "ONCE": 11,
    "DOCE O MÁS": 12
}

mapa_rango = {
    "1 A 2": 2,
    "3 A 4": 4,
    "5 A 6": 6,
    "7 A 8": 8,
    "9 O MÁS": 9,
    "9 O MAS": 9
}

numero = col.map(mapa_numeros)
numero = numero.fillna(col.replace(mapa_rango))

numero = pd.to_numeric(numero, errors="coerce")

cat = pd.Series("Sin información", index=col.index)

cat[numero.between(1, 2, inclusive="both")] = "1-2"
cat[numero.between(3, 4, inclusive="both")] = "3-4"
cat[numero.between(5, 6, inclusive="both")] = "5-6"
cat[numero.between(7, 8, inclusive="both")] = "7-8"
cat[numero >= 9] = "9+"

conteo = cat.value_counts(dropna=False)
prop = conteo / conteo.sum() * 100

orden = ["1-2", "3-4", "5-6", "7-8", "9+", "Sin información"]
conteo = conteo.reindex(orden, fill_value=0)
prop = prop.reindex(orden, fill_value=0)

plt.figure(figsize=(8,4))
ax = prop.plot(kind="barh")
plt.title("Personas en el hogar (%)")
plt.xlabel("Porcentaje (%)")
plt.ylabel("Categoría")

for i, c in enumerate(prop.index):
    pct = prop.loc[c]
    n = conteo.loc[c]
    ax.text(pct, i, f"  {n} ({pct:.1f}%)", va="center", fontsize=9)

ax.set_xlim(0, prop.max() * 1.2 if prop.max() > 0 else 1)
plt.tight_layout()
plt.show()
```
