### Efecto del estrato en las áreas distintas

```python
columna_estrato = 'FAMI_ESTRATOVIVIENDA'
estratos = ['Sin Estrato', 'Estrato 1', 'Estrato 2', 'Estrato 3', 'Estrato 4', 'Estrato 5', 'Estrato 6']
areas = ['PUNT_MATEMATICAS', 'PUNT_INGLES', 'PUNT_SOCIALES_CIUDADANAS', 'PUNT_C_NATURALES', 'PUNT_LECTURA_CRITICA']
labels_areas =["Puntaje en matemáticas", "Puntaje en Inglés", "Puntaje en ciencias sociales", "Puntaje en ciencias naturales", "Puntaje en lectura crítica"]

colores = {
    "PUNT_MATEMATICAS": "#42fff9",
    "PUNT_INGLES": "#fc9840",
    "PUNT_SOCIALES_CIUDADANAS": "#4eff4e",
    "PUNT_C_NATURALES": "#ff714a",
    "PUNT_LECTURA_CRITICA": "#b15fff"
}

means = df.groupby(columna_estrato)[areas].mean()
means = means.reindex(estratos)

x = np.arange(len(means.index))
n = len(areas)
group_w = 0.65
w = group_w / n


fig, ax = plt.subplots(figsize=(14, 5))

for i, area in enumerate(areas):
    ax.bar(x+(i-(len(areas)-1)/2)*w, means[area].values, width=w*0.85, label = area)

ax.set_xticks(x)
ax.set_xticklabels(means.index.astype(str))


ax.set_title('Áreas más afectadas por condición económica')
ax.set_xlabel('Estrato')
ax.set_ylabel('Puntaje promedio')
ax.legend(ncol=len(areas), bbox_to_anchor=(0.5, -0.15), loc='upper center', frameon=False)

plt.tight_layout()
plt.show()


```

### Acceso a internet y desempeño en idiomas

```python
df["DESEMP_INGLES"].value_counts()
```

```python
internet_columna = 'FAMI_TIENEINTERNET'
columna_nivel_ing = 'DESEMP_INGLES'
columna_ingles = "PUNT_INGLES"
columna_global = "PUNT_GLOBAL"

niveles = ["A-", "A1", "A2", "B1", "B+"]

internet = df[internet_columna].astype(str).str.lower().str.strip()
con_internet = internet.isin(["si"])

#Grafico 1

df2=df.copy()
df2["Internet"] = np.where(con_internet, "Con internet", "Sin internet")

print(df2["Internet"].value_counts())

fig, axes = plt.subplots(1, 2, figsize=(16, 6))
fig.suptitle('Acceso a internet y desempeño en idiomas', fontsize=16, fontweight='bold')

ax1 = axes[0]

conteos = (df2.groupby([columna_nivel_ing, "Internet"]).size().unstack(fill_value=0))

conteos = conteos.reindex(niveles)

x = np.arange(len(conteos.index))
w = 0.5

ax1.bar(x - w/2, conteos.get("Con internet", 0), width = w*0.7, label = "Con internet", color = "#03eaf2")
ax1.bar(x + w/2, conteos.get("Sin internet", 0), width = w*0.7, label = "Sin internet", color = '#ff5a3d')
        
ax1.set_title('Distribución de niveles de ingles por acceso a internet')
ax1.set_xticks(x)
ax1.set_xticklabels(conteos.index)
ax1.set_ylabel('Cantidad de estudiantes')
ax1.legend(frameon=False)

#Grafico 2
ax2 = axes[1]


m = df2[[columna_ingles, columna_global, 'Internet']].dropna()
for grp, col in [('Con internet', "#03eaf2"), ('Sin internet', '#ff5a3d')]:
    sub = m[m['Internet'] == grp]
    ax2.scatter(sub[columna_ingles], sub[columna_global], s=30, alpha=0.45, label=grp, c=col)

ax2.set_title('Puntaje inglés vs puntaje global')
ax2.set_xlabel('Punt. Inglés')
ax2.set_ylabel('Punt. Global')
ax2.legend(frameon=False)

plt.tight_layout()
plt.show()



```
