### Efecto del estrato en las áreas distintas

```python
columna_estrato = 'FAMI_ESTRATOVIVIENDA'
estratos = ['Sin Estrato', 'Estrato 1', 'Estrato 2', 'Estrato 3', 'Estrato 4', 'Estrato 5', 'Estrato 6']
areas = ['PUNT_MATEMATICAS', 'PUNT_INGLES', 'PUNT_SOCIALES_CIUDADANAS', 'PUNT_C_NATURALES', 'PUNT_LECTURA_CRITICA']
labels_areas =["Puntaje en matemáticas", "Puntaje en Inglés", "Puntaje en ciencias sociales", "Puntaje en ciencias naturales", "Puntaje en lectura crítica"]

colores = {
    "PUNT_MATEMATICAS": "#42fff9",
    "PUNT_INGLES": "#fc9840",
    "PUNT_SOCIALES_CIUDADANAS": "#4eff4e",
    "PUNT_C_NATURALES": "#ff714a",
    "PUNT_LECTURA_CRITICA": "#b15fff"
}

means = df.groupby(columna_estrato)[areas].mean()
means = means.reindex(estratos)

x = np.arange(len(means.index))
n = len(areas)
group_w = 0.65
w = group_w / n


fig, ax = plt.subplots(figsize=(14, 5))

for i, area in enumerate(areas):
    ax.bar(x+(i-(len(areas)-1)/2)*w, means[area].values, width=w*0.85, label = labels_areas[i], color= colores[area])

ax.set_xticks(x)
ax.set_xticklabels(means.index.astype(str))


ax.set_title('Áreas más afectadas por condición económica')
ax.set_xlabel('Estrato')
ax.set_ylabel('Puntaje promedio')
ax.legend(ncol=len(areas), bbox_to_anchor=(0.5, -0.15), loc='upper center', frameon=False)

plt.tight_layout()
plt.show()


```

### Acceso a internet y desempeño en idiomas

```python
df["DESEMP_INGLES"].value_counts()
```

```python
internet_columna = 'FAMI_TIENEINTERNET'
columna_nivel_ing = 'DESEMP_INGLES'
columna_ingles = "PUNT_INGLES"
columna_global = "PUNT_GLOBAL"

niveles = ["A-", "A1", "A2", "B1", "B+"]

internet = df[internet_columna].astype(str).str.lower().str.strip()
con_internet = internet.isin(["si"])

#Grafico 1

df2=df.copy()
df2["Internet"] = np.where(con_internet, "Con internet", "Sin internet")

print(df2["Internet"].value_counts())

fig, axes = plt.subplots(1, 2, figsize=(16, 6))
fig.suptitle('Acceso a internet y desempeño en idiomas', fontsize=16, fontweight='bold')

ax1 = axes[0]

conteos = (df2.groupby([columna_nivel_ing, "Internet"]).size().unstack(fill_value=0))

conteos = conteos.reindex(niveles)

x = np.arange(len(conteos.index))
w = 0.5

ax1.bar(x - w/2, conteos.get("Con internet", 0), width = w*0.7, label = "Con internet", color = "#03eaf2")
ax1.bar(x + w/2, conteos.get("Sin internet", 0), width = w*0.7, label = "Sin internet", color = '#ff5a3d')
        
ax1.set_title('Distribución de niveles de ingles por acceso a internet')
ax1.set_xticks(x)
ax1.set_xticklabels(conteos.index)
ax1.set_ylabel('Cantidad de estudiantes')
ax1.legend(frameon=False)

#Grafico 2
ax2 = axes[1]


m = df2[[columna_ingles, columna_global, 'Internet']].dropna()
for grp, col in [('Con internet', "#03eaf2"), ('Sin internet', '#ff5a3d')]:
    sub = m[m['Internet'] == grp]
    ax2.scatter(sub[columna_ingles], sub[columna_global], s=30, alpha=0.45, label=grp, c=col)

ax2.set_title('Puntaje inglés vs puntaje global')
ax2.set_xlabel('Punt. Inglés')
ax2.set_ylabel('Punt. Global')
ax2.legend(frameon=False)

plt.tight_layout()
plt.show()



```

### Dispersión de puntajes por tipo de colegio y por zona

```python
df["COLE_NATURALEZA"]
```

```python
tipo_colegio = "COLE_NATURALEZA"
zona_colegio = "COLE_AREA_UBICACION"

puntaje = "PUNT_GLOBAL" # PUNT_MATEMATICAS - PUNT_INGLES - PUNT_SOCIALES_CIUDADANAS PUNT_C_NATURALES - PUNT_LECTURA_CRITICA - PUNT_GLOBAL

areas =["PUNT_MATEMATICAS", "PUNT_INGLES", "PUNT_SOCIALES_CIUDADANAS", "PUNT_C_NATURALES", "PUNT_LECTURA_CRITICA"]
temporal = df[[tipo_colegio, zona_colegio]+ areas].dropna().copy()
temporal[tipo_colegio] = temporal[tipo_colegio].astype(str).str.lower().str.strip()
temporal[zona_colegio] = temporal[zona_colegio].astype(str).str.lower().str.strip()

def mapa_tipo(colegio):
    if "no oficial" in colegio:
        return "Privado"
    return "Oficial"

def mapa_zona(colegio):
    if "rural" in colegio:
        return "Rural"
    return "Urbano"

temporal["Tipo"] = temporal[tipo_colegio].map(mapa_tipo)
temporal["Zona"] = temporal[zona_colegio].map(mapa_zona)
temporal["Grupo"] = temporal["Tipo"] + " - " + temporal["Zona"]

temporal.drop(tipo_colegio, axis=1,inplace=True)
temporal.drop(zona_colegio, axis=1, inplace=True)

orden = ["Oficial - Urbano", "Oficial - Rural", "Privado - Urbano", "Privado - Rural"]

colores = colores = {
    "Oficial - Urbano": "#42fff9",
    "Oficial - Rural": "#fc9840",
    "Privado - Urbano": "#4eff4e",
    "Privado - Rural": "#b15fff",
}


tabla = (temporal.groupby("Grupo")[areas].mean().reindex(orden))
tabla = tabla.T
 

import plotly.express as px

fig = px.imshow(tabla,
                text_auto = ".1f",
                aspect = "equal",
                labels=dict(x="Grupo", y="Área", color= '#RdYlGn'),
                x=["Oficial - Urbano", "Oficial - Rural", "Privado - Urbano", "Privado - Rural"],
                y=["Matemáticas", "Inglés", "Ciencias Sociales", "Ciencias Naturales", "Lectura Crítica"]
               )
fig.update_xaxes(side="top")
fig.update_layout(
    title="Puntaje promedio por grupo y área",
    width=700,
    height=800,
    margin=dict(l=80, r=40, t=80, b=60),
    coloraxis_colorbar=dict(
        title="Puntaje Promedio",
        len=0.65,
        thickness=30,
        x=1.3
    )
)



fig.show()



```

## Puntajes vs variables socioeconómicas

```python
import pandas as pd
import matplotlib.pyplot as plt

def bivariado_puntaje_vs_categoria(
    dft,
    puntaje_col: str,
    cat_col: str,
    order: list | None = None,
    top_n: int | None = None,
    other_label: str = "OTROS",
    missing_label: str = "Sin información",
    drop_missing_cat: bool = False,
    figsize=(10, 6),
):
    temporal = dft[[cat_col, puntaje_col]].copy()

    temporal[cat_col] = (
        temporal[cat_col].astype("string").str.strip().fillna(missing_label)
    )

    temporal[puntaje_col] = pd.to_numeric(
        temporal[puntaje_col].astype("string").str.strip(),
        errors="coerce"
    )

    temporal = temporal.dropna(subset=[puntaje_col])

    if drop_missing_cat:
        temporal = temporal[temporal[cat_col] != missing_label]

    if top_n is not None:
        vc = temporal[cat_col].value_counts()
        top_cats = vc.head(top_n).index
        temporal[cat_col] = temporal[cat_col].where(
            temporal[cat_col].isin(top_cats),
            other_label
        )

    if order is None:
        cats = temporal[cat_col].value_counts().index.tolist()
    else:
        present = temporal[cat_col].unique().tolist()
        extras = [c for c in present if c not in order]
        cats = [c for c in order if c in present] + extras

    data = [
        temporal.loc[temporal[cat_col] == c, puntaje_col].dropna().values
        for c in cats
    ]

    fig, ax = plt.subplots(figsize=figsize, constrained_layout=True)
    ax.boxplot(data, labels=cats, vert=False, showfliers=False)
    ax.set_title(f"{puntaje_col} vs {cat_col}")
    ax.set_xlabel(puntaje_col)
    ax.set_ylabel(cat_col)
    fig.subplots_adjust(left=0.35)
    plt.show()

    resumen = (temporal.groupby(cat_col)[puntaje_col]
               .agg(n="count", mean="mean", median="median", std="std", min="min", max="max")
               .reindex(cats))

    return resumen
```

## Categóricas vs Categóricas

```python
def cat_vs_cat_porcentaje(dft, fila_col, col_col, order_fila=None, order_col=None,
                          fila_missing="Sin información", col_missing="Sin información",
                          top_n_col=None, other_label="OTROS", figsize=(10,5)):

    temporal = dft[[fila_col, col_col]].copy()

    temporal[fila_col] = temporal[fila_col].astype("string").str.strip().fillna(fila_missing)
    temporal[col_col]  = temporal[col_col].astype("string").str.strip().fillna(col_missing)

    if top_n_col is not None:
        vc = temporal[col_col].value_counts()
        top = vc.head(top_n_col).index
        temporal[col_col] = temporal[col_col].where(temporal[col_col].isin(top), other_label)

    ct = pd.crosstab(temporal[fila_col], temporal[col_col], normalize="index") * 100

    if order_fila is not None:
        ct = ct.reindex(order_fila).fillna(0)
    if order_col is not None:
        ct = ct.reindex(columns=order_col).fillna(0)

    ax = ct.plot(kind="bar", stacked=True, figsize=figsize)
    ax.set_title(f"{col_col} vs {fila_col} (porcentaje por {fila_col})")
    ax.set_xlabel(fila_col)
    ax.set_ylabel("Porcentaje (%)")
    plt.xticks(rotation=0)
    plt.legend(title=col_col, bbox_to_anchor=(1.02, 1), loc="upper left")
    plt.tight_layout()
    plt.show()

    return ct
```
